Neste projeto ser치 realizado uma an치lise envolvendo os diametros e os pesos de laranjas e toranjas

Importando a biblioteca Numpy para o desenvolvimento do projeto

```python
import numpy as np
```

Importando os dados necess치rios para o estudo

```python
url = 'https://raw.githubusercontent.com/ThiagoFsm/projetos-alura/refs/heads/main/Laranja%20e%20Toranja/citrus.csv'
```

```python
dado = np.loadtxt(url, delimiter=',', skiprows=1, usecols=np.arange(1, 3, 1))
```

Iniciando o tratamento dos dados

```python
dado_transposto = dado.T
dado_transposto
```

Separando os dados entre Laranjas e Toranjas

```python
diametro_laranja = dado_transposto[0, :5000]
diametro_toranja = dado_transposto[0, 5000:]
peso_laranja = dado_transposto[1, :5000]
peso_toranja = dado_transposto[1, 5000:]
```

```python
print(diametro_laranja)
print(peso_laranja)
```

```python
print(diametro_toranja)
print(peso_toranja)
```

Importando a bibilioteca Matplotlib para realizar as an치lises dos dados

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(diametro_laranja, peso_laranja)
plt.plot(diametro_toranja, peso_toranja)
plt.legend(['Laranja', 'Toranja'])
plt.xlabel('Diametro')
plt.ylabel('Peso')
plt.title('Peso x Diametro')
plt.show()
```

Iniciando o tratamento para uma regress칚o linear em ambas as frutas

```python
Y_laranja = peso_laranja
Y_toranja = peso_toranja
X_laranja = diametro_laranja
X_toranja = diametro_toranja
n_laranja = np.size(X_laranja)
n_toranja = np.size(X_toranja)
```

```python
a_laranja = (n_laranja*np.sum(X_laranja*Y_laranja)-np.sum(X_laranja)*np.sum(Y_laranja))/(n_laranja*np.sum(X_laranja**2)-np.sum(X_laranja)**2)
a_toranja = (n_toranja*np.sum(X_toranja*Y_toranja)-np.sum(X_toranja)*np.sum(Y_toranja))/(n_toranja*np.sum(X_toranja**2)-np.sum(X_toranja)**2)
b_laranja = np.mean(Y_laranja)-a_laranja*np.mean(X_laranja)
b_toranja = np.mean(Y_toranja)-a_toranja*np.mean(X_toranja)
```

Regress칚o linear completa para uma breve an치lise das Laranjas

```python
plt.plot(X_laranja, Y_laranja, 'blue')
plt.plot(X_laranja, a_laranja*X_laranja+b_laranja, 'black')
```

Regress칚o linear completa para uma breve an치lise das Toranjas

```python
plt.plot(X_toranja, Y_toranja, 'orange')
plt.plot(X_toranja, a_toranja*X_toranja+b_toranja, 'black')
```
